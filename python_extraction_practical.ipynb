```python
def read_config(file_path:str)->dict:
    with open(file_path, encoding="utf8") as json_file:
        data = json.load(json_file)
    return data
```

```python
config = read_config(file_path = "keyword_chatbot.json")
intents = config['bocgi']['intents']
```

```python
def extract_config(intents:list)->list:
    results = []
    for intent in intents:
        result = [intent['id'], intent['name'], intent['remarks']]
        results.append(result)
    return results
```

```python
results = extract_config(intents = intents)
```

```python
def save_csv(data:list, file_name:str='results.csv'):
  
    fields = ['id','name','remarks']

    # opening the csv file in 'w+' mode 
    file = open(file_name, 'w', encoding="utf8") 
  
    # writing the data into the file 
    with file:     
        write = csv.writer(file, delimiter=',') 
        write.writerow(fields)
        write.writerows(data)
```

```python
save_csv(data = results, file_name = "abc.csv")
```
